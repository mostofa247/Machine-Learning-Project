###Question 1 Solution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px



url = "https://github.com/phitronio/Python-for-ML/raw/main/final-employee-ds.csv"
df = pd.read_csv(url)
TrainHours = df["TrainingHours"].to_numpy().astype(float)
TrainMean = np.mean(TrainHours)
Standard_deviation = np.std(TrainHours)
print(TrainMean)
print(Standard_deviation)
```

###Question 2 Solution

```python
df.isnull().sum()
```

```python
dep_Salary_median = df.groupby("Department")["Salary"].transform("median")
dep_Salary_median
df["Salary"]=df["Salary"].fillna(dep_Salary_median)
df.isnull().sum()
```

```python
filter_data = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]
print(filter_data["Salary"].sum())
```

###Question 3 Solution


```python
searching_data = df[((df["Department"]=="IT") | (df["Department"] == "Finance")) & df["Age"].between(30,45) & (df["PerformanceScore"] > 88)]
searching_data = searching_data.sort_values(by="Salary", ascending=False)
result = searching_data[["Name", "Department", "Salary", "PerformanceScore"]]
print(result)
```

###Question 4 Solution

```python
df["SalaryPerHour"] = df["Salary"] / (df["WorkHoursPerWeek"] * 4)
topSalary5 = df.sort_values("SalaryPerHour", ascending=False).head(5)
answer = topSalary5[["Name", "SalaryPerHour"]]
print(answer)
```
