```python
from sklearn.preprocessing import LabelEncoder


df_heart = pd.read_csv("heart_disease.csv")
df_heart.head()
```

```python
category_cols = ["Gender", "Exercise Habits", "Smoking", "Family Heart Disease", "Diabetes",
                 "High Blood Pressure", "Low HDL Cholesterol", "High LDL Cholesterol", "Alcohol Consumption",
                 "Stress Level", "Sugar Consumption", "Heart Disease Status"]

for c in category_cols:
  plt.figure(figsize=(5,4))
  df_heart[c].value_counts().plot(kind="bar")
  plt.title(f"Value counts for {c}")
  plt.ylabel("Count")
  plt.tight_layout()
  plt.show()
```

```python
df_heart.info()
```

```python
le = LabelEncoder()
df_heart["Heart Disease Status"] = le.fit_transform(df_heart["Heart Disease Status"])
df_heart["Gender"] = le.fit_transform(df_heart["Gender"])
df_heart["Smoking"] = le.fit_transform(df_heart["Smoking"])
df_heart["Family Heart Disease"] = le.fit_transform(df_heart["Family Heart Disease"])
```

```python
df_heart.head(10)
```

```python
cat_cols = ["Exercise Habits", "Alcohol Consumption"]

df_heart_encoded = pd.get_dummies(
    df_heart,
    columns= cat_cols,
    dtype=int
)
```

```python
df_heart_encoded.head()
```

```python
plt.figure(figsize=(5,4))
df_heart["Heart Disease Status"].value_counts().plot(kind="bar")
plt.title("Value counts for Heart Disease Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler, MinMaxScaler, robust_scale
from sklearn.preprocessing import OneHotEncoder


df_heart_encoded = pd.get_dummies(df_heart_encoded, drop_first=True)

target_col = "Heart Disease Status"

x = df_heart_encoded.drop(columns=[target_col])
y = df_heart_encoded[target_col]

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)

scaler_sd = StandardScaler()
x_train_std = scaler_sd.fit_transform(x_train)
x_test_std = scaler_sd.transform(x_test)

scaler_mm = MinMaxScaler()
x_train_mm = scaler_mm.fit_transform(x_train)
x_test_mm = scaler_mm.transform(x_test)

x_train_std_df = pd.DataFrame(x_train_std, columns=x_train.columns, index=x_train.index)
x_test_std_df = pd.DataFrame(x_test_std, columns=x_test.columns, index=x_test.index)


x_train_mm_df = pd.DataFrame(x_train_mm, columns=x_train.columns, index=x_train.index)
x_test_mm_df = pd.DataFrame(x_test_mm, columns=x_test.columns, index=x_test.index)


display(x_train_std_df.head())
display(x_test_std_df.head())

display(x_train_mm_df.head())
display(x_test_mm_df.head())
```
